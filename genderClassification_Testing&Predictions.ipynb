```python
import os
import pickle
import numpy as np
import pandas as pd
import warnings
warnings.simplefilter('ignore')

model_location = r'C:\Users\Sagar Mishra\jupyter notebooks\genderClassification.sav'
loaded_model = pickle.load(open(model_location, 'rb'))

test_dataset = r'D:\Datasets\gender-names\name_gender.csv'
data_test = pd.read_csv(test_dataset)
data_test.head(10)
```

```python
#Preparing the Data:
for i, gender in enumerate(data_test['Gender']):
  if gender.lower() == 'm':
    data_test['Gender'][i] = 1
    data_test['Score'][i] = float(data_test["Score"][i])
  if gender.lower() == 'f':
    data_test['Gender'][i] = 0
    data_test['Score'][i] = 1.0 - float(data_test["Score"][i])

data_test['length'] = data_test.apply(lambda row: len(row.Name), axis=1)
vowels = 'aeiou'
data_test['no_of_vowels'] = 0
data_test['percentage_of_vowels'] = 0.0
data_test['ends_with_vowels'] = 0

for i, names in enumerate(data_test['Name']):
  count_vowels = 0
  for alpha in names.lower():
    if alpha in vowels:
      count_vowels += 1
  data_test['no_of_vowels'][i] = count_vowels
  data_test['percentage_of_vowels'][i] = count_vowels/(len(names))
  if names[len(names)-1] in vowels:
    data_test['ends_with_vowels'][i] = 1
  else:
    data_test['ends_with_vowels'][i] = 0
    
maxLength = data_test['length'].max()

initial = 'abcdefghijklm'
final = 'nopqrstuvwxyz'
dict_alpha = {}
for i in (initial+final):
  if i in initial:
    dict_alpha[i] = float(ord(i) -96 -14)
  if i in final:
    dict_alpha[i] = float(ord(i) -96 -13)
dict_alpha["'"] = 0.0
dict_alpha["-"] = 0.0
dict_alpha[","] = 0.0
dict_alpha[" "] = 0.0
dict_alpha["."] = 0.0
dict_alpha["\\"] = 0.0
dict_alpha["/"] = 0.0
dict_alpha["("] = 0.0
dict_alpha[")"] = 0.0
dict_alpha["["] = 0.0
dict_alpha["]"] = 0.0
dict_alpha["{"] = 0.0
dict_alpha["}"] = 0.0

for i in range(10):
  dict_alpha[str(i)] = 0.0


for i in range(22):
  data_test["{}st_character".format(i)] = 0

for i, names in enumerate(data_test['Name']):
  for j in range(maxLength):
    if j < len(names):
      data_test["{}st_character".format(j)][i] = dict_alpha[names[j].lower()]
    else:
      data_test["{}st_character".format(j)][i] = 0.0
    
data_test = data_test.sample(frac=1).reset_index(drop=True)
y = data_test['Gender']
data_test.drop(labels="Gender", axis=1, inplace=True)

data_test.head(10)
```
